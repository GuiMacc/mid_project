```python
import numpy as np
import pandas as pd

#basic libraries to explore data
```

```python
game_data=pd.read_csv('../Data/games-data.csv')
game_data.head()
```

```python
#changing column names into ones that give better understanding of the data

game_data.rename(columns = {'name' : 'game', 'platform' : 'console', 'r-date' : 'release_date',
                            'score' : 'critic_score', 'user score' : 'user_score',
                            'developer' : 'dev', 'players' : 'player_support','critics' : 'num_critics', 'users' : 'num_users'}, inplace = True )
game_data.head()
```

```python
data=game_data.copy()

#working on a copy to make it easier to correct mistakes we make down the line

display(data.head())
```

```python
#doing an initial check on both null values and column types

display(data.info())
```

```python
display(data['num_users'].value_counts())
display(data['user_score'].unique())
display(data['num_critics'].unique())
display(data['critic_score'].unique())
```

```python
#Some games have 0 critics or users and there's even a few with no user score,
#so the first thing we need to do is to get rid of those,
#since my main focus is analysing most other columns based on scores.
#For simplicity sake I'm assuming that a review score left by 0 people is fraudulent.
```

```python
#let's check some rows with either 0 critics or 0 users

display(data[data['num_critics']==0]) 
display(data[data['num_users']==0])
```

```python
#let's now drop those unwanted rows and check column types

data=data[(data['user_score'] != 'tbd') & (data['num_users'] != 0) & (data['num_critics'] != 0)]
display(data.info())
```

```python
#checking for null values

data.isnull().sum()
```

```python
#checking the values of the column with nulls to see if there's an easy fill option

data['player_support'].unique()
```

```python
data[data['player_support'].isnull()]
```

```python
#it seems that some rows don't have info on the support type, which is a bit annoying but we can use it for the null values

data['player_support'].fillna('No info', inplace=True)
display(data['player_support'].unique())
display(data.isnull().sum())
```

```python
#Since we no longer have nulls let's go ahead and apply a quick fix to one of our main columns regarding exploration.
#While the critic score comes on a 0-100 scale, the user score is on a 0-10.
#Let's simply multiply the user score by 10 and cast it as an int so we can more easily compare the two score types.

def fix_user_score(x):
    y = float(x) * 10
    return int(y)

data['user_score']=data['user_score'].apply(fix_user_score)
data['user_score'].unique()
```

```python
#Now that both scores are on the same scale, let's add a new column which allows us to compare critic score with user score
#There might be games where the critics liked it, but it didn't actually appeal to the general audiance or vice-versa
#That could be really interesting to explore if there's a console that stands out or even the type of player support
#We'll subtract the critic score FROM the user score
#A positive value indicates users have a better score, and a negative value indicates critics have a better score

data['score_diff']=data['user_score']-data['critic_score']
data.head()
```

```python
#all we have left before we start analysing the categorical columns
#is change the release column to datetime and extracting month and year as integers

data['release_date']= pd.to_datetime(data['release_date'])
data['release_year']=pd.DatetimeIndex(data['release_date']).year
data['release_month']=pd.DatetimeIndex(data['release_date']).month
data.head()
```

```python
#let's rearrange the columns a bit

data=data[['game', 'console', 'dev', 'genre', 'player_support', 'critic_score', 'user_score', 'score_diff',
           'num_critics', 'num_users' , 'release_date', 'release_month', 'release_year']]
data.head()
```

```python
#Now that we seem to have the initial preparation out of the way
#it's time to go through each categorical column, check the amounth of values
#and, if necessary, combine some values

data.info()
```

```python
#getting a dataframe with only the columns we want to check
#will use this as the input for the functions without ever changing it
#if my function ever returns me strange things, the original input is still there and i don't need to clear any variables

cat_data=data.select_dtypes(include='object')
cat_data.drop('game', axis=1, inplace=True)
cat_data.head()
```

```python
for col in cat_data.columns:
    display(cat_data[col].nunique())
```

```python
#we definitely need to clean all collumns, but dev and genre especially
#let's start with the smaller ones though
```

```python
#let's check the unique values for the console column

cat_data['console'].unique()
```

```python
#if we really want to analyse individual consoles we can't do much,
#but we can create another column if we want to compare the main console producers

def console_groupping(row):
    if "PlayStation" in row or row == "PSP":
        return "PS"
    elif "Xbox" in row:
        return "Xbox"
    elif row == "Dreamcast":
        return "Sega"
    elif row == "Stadia":
        return "Google"
    elif row == "PC":
        return "PC"
    else:
        return "Nintendo"
    
data['console_brand']=cat_data['console'].apply(console_groupping)
```

```python
display(data['console'].value_counts())
display(data['console_brand'].value_counts())
```

```python
#now we can have a much broader plot with the new column or a more narrow one with the old column

#let's move onto cleaning the player_support column
```

```python
display(cat_data['player_support'].unique())
display(cat_data['player_support'].value_counts())
```

```python
cat_data[cat_data['player_support']=='4  Online']
```

```python
#we have way too many values here, but we'll try to simplify them into 4 (counting "no info"):
#singleplayer, co-op, and multiplayer
#the biggest problem, is where to draw the line
#any games with online could be counted as multiplayer, howevere there are online co-op games
#a game with up to 4 players could either be an online co-op or a regular multiplayer
#to keep it simple, even if it might not be 100% exact, i'll aggregate any games with a maximum of 4 players into co-op
```

```python
def fix_player_count(row):
    if row == "No info":
        return "No info"
    elif row == "1 Player" or row == "No Online Multiplayer":
        return "Singleplayer"
    elif row=="1-2 " or row=="1-3 " or row=="1-4 " or row=="2 " or row=="Up to 3 " or row=="Up to 4 " or row=="2  Online"  or row=="3  Online"  or row=="4  Online" :
        return "Co-op"
    else:
        return "Multiplayer"
```

```python
data['player_support']=cat_data['player_support'].apply(fix_player_count)
display(data['player_support'].value_counts())
```

```python
#looks way nicer now
#let's move onto the first nightmare: genre
```

```python
cat_data['genre'].unique()
```

```python
#from the looks of it, most games have multiple genres, which makes sense
#however we don't want to have a massive plot with every single genre combination
#so my initial idea is to get the main genre, and potentially a secondary genre for the games that have it
```

```python
def genre_cleaning(row):
    g=row.split(",")
    return g[0]

data['main_genre']=cat_data['genre'].apply(genre_cleaning)
print(data['main_genre'].nunique())
display(data['main_genre'].value_counts().head(15))
```

```python
#still ended up with 60+ genres, so let's group the least represented ones into "Other"

data['main_genre']= np.where(data['main_genre'].isin(['Action', 'Action Adventure', 'Role-Playing', 'Sports', 'Strategy', 'Miscellaneous', 'Adventure', 'Driving', 'Simulation']) , data['main_genre'], 'Other')
data['main_genre'].value_counts()
```
